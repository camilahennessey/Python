## Problem 1: Environmental Concerns with Hydraulic Fracturing

### Step 1: Quantifying Water Usage: 
The code begins by calculating the total water pumped (TotalWaterPumped_GAL) and the cumulative production (CumProd_BOE). This fundamental step quantifies the scale of water usage in hydraulic fracturing operations, setting the stage for deeper analysis.

```python
# Calculate total water pumped and cumulative production
total_water_pumped = metadata_cleaned['TotalWaterPumped_GAL'].sum()
total_cumulative_production = metadata_cleaned['CumProd_BOE'].sum()

# Calculate the overall water to oil ratio
overall_water_to_oil_ratio = total_water_pumped / total_cumulative_production
print(f"Overall Water to Oil Ratio: {overall_water_to_oil_ratio}")
```

### Step 2: Operator-Specific Water Usage Analysis

We are now going to identify how much water each operator uses for fracking, highlighting discrepancies and allowing for targeted improvements in water use efficiency. By calculating the overall water to oil ratio, the analysis highlights the efficiency of water usage relative to oil production. This ratio is a key indicator of the environmental impact of fracking operations, particularly in terms of water use.

```python
# Calculate total and average water usage for fracturing by operator
metadata_cleaned['TotalWaterPumped_GAL'] = pd.to_numeric(metadata_cleaned['TotalWaterPumped_GAL'], errors='coerce')
total_water_by_operator = metadata_cleaned.groupby('ENVOperator')['TotalWaterPumped_GAL'].sum()
average_water_by_operator = metadata_cleaned.groupby('ENVOperator')['TotalWaterPumped_GAL'].mean()

# Output the total and average water usage
print("Total Water Usage by Operator (GAL):")
print(total_water_by_operator)
print("\nAverage Water Usage by Operator (GAL):")
print(average_water_by_operator)
```

### Step 3: Water Usage Efficiency

This segment focuses on efficiency by comparing water use to oil production across operators, identifying leaders and laggards in water use efficiency. The analysis then drills down to calculate total and average water usage by operator. This granular view identifies which operators are using water more efficiently and which may require improvements in their operations to minimize environmental impact.

```python
# Calculate water usage per BOE produced for each operator
metadata_cleaned['WaterUsagePerBOE'] = metadata_cleaned['TotalWaterPumped_GAL'] / metadata_cleaned['CumProd_BOE']

# Identify operators with the highest and lowest water usage per BOE
highest_water_usage_per_boe = metadata_cleaned.sort_values(by='WaterUsagePerBOE', ascending=False).head()
lowest_water_usage_per_boe = metadata_cleaned.sort_values(by='WaterUsagePerBOE').head()

print("Operators with Highest Water Usage Per BOE:")
print(highest_water_usage_per_boe[['ENVOperator', 'WaterUsagePerBOE']])

print("\nOperators with Lowest Water Usage Per BOE:")
print(lowest_water_usage_per_boe[['ENVOperator', 'WaterUsagePerBOE']])
```

### Step 4: Gas to Water Production Ratio Analysis

Although not directly stated as a step for Problem 1, this analysis complements the water use efficiency evaluation by offering insights into the balance between gas production and water use, further informing environmental sustainability efforts. The code further analyzes the gas to water production ratio, providing insights into the relative efficiency of gas production versus water usage over time. This analysis is extended to identify yearly trends, offering a long-term perspective on water use efficiency in relation to gas production.

```python
# Calculate the ratio of gas production to water production to understand the efficiency
production_data['GasToWaterRatio'] = production_data['GasProd_MCF'] / production_data['WaterProd_BBL']
gas_to_water_ratio_trends = production_data.groupby('ProducingMonth')['GasToWaterRatio'].mean()

# Assuming `ProducingMonth` is in a datetime format; if not, convert it
production_data['ProducingMonth'] = pd.to_datetime(production_data['ProducingMonth'])
gas_to_water_ratio_trends_yearly = production_data.resample('Y', on='ProducingMonth')['GasToWaterRatio'].mean()

print("Yearly Gas to Water Production Ratio Trends:")
print(gas_to_water_ratio_trends_yearly.dropna())
```
